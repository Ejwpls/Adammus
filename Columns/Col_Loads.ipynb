```python
import xlwings as xw
import pandas as pd

file_path = r"C:\Users\edber\Desktop\HTK_COLUMN DESIGN_MODULE 2.0.xlsm"
wb = xw.Book(file_path)
sheet = wb.sheets['LOADING']

tbl_ColLoads = sheet.range('tbl_ColLoads').options(pd.DataFrame, header=0, index=False, expand='table').value

#put to dataframe 
df = pd.DataFrame(tbl_ColLoads)

#change df column names to excel table 'tbl_ColLoads' column names
df.columns = ['GUID','Story','Column','Unique Name', 'Load Case/Combo','Station','P','V2','V3','T','M2','M3']
```

```python
#df groupby Column and Story
df_grouped = df.groupby(['Column','Story','Load Case/Combo'])

#Show data only the max and min of Station for each group
df_grouped.apply(lambda x: x[(x['Station'] == x['Station'].max()) | (x['Station'] == x['Station'].min())])
```
