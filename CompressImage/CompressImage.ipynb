```python
import os
from PIL import Image

# Function to compress the image
def compress_image(image_file):
    # Accessing the image file
    filepath = os.path.join(os.getcwd(), image_file)
    
    # Opening the file
    image = Image.open(filepath)
    
    # Checking if the image size is greater than 1MB
    if os.path.getsize(filepath) > 1e5:
        # Compressing the image
        compressed_image = image.resize((image.width // 2, image.height // 2))
        # Saving the compressed image
        compressed_filepath = os.path.join(filepath)
        compressed_image.save(compressed_filepath, optimize=True, quality=85)
        while os.path.getsize(filepath) > 1e6:
            compressed_image = compressed_image.resize((compressed_image.width // 2, compressed_image.height // 2))
            compressed_image.save(compressed_filepath, optimize=True, quality=85)
        
        return compressed_filepath
    
    return filepath

# Function to compress all images in a folder until their size is below 1MB
def compress_images_in_folder(folder_path):
    # Getting the list of files in the folder
    files = os.listdir(folder_path)
    
    # Compressing each image file
    for file in files:
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
            image_file = os.path.join(folder_path, file)
            compressed_filepath = compress_image(image_file)
            
            if os.path.getsize(compressed_filepath)> 1e6:
                print(f"Compressed image saved as: {compressed_filepath}")
            else:
                print(f"{os.path.getsize(compressed_filepath)}: {image_file}")

```
